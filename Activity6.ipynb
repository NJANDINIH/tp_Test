#DATA VISUALIZATION
Creating a histogram using foot_ball["Age"] column: average(mean) = 25.789474 standard deviation(std) = 4.377354 total_rows(size) = 571

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(loc=10, scale=5.6, size=571)
plt.hist(data,bins=10, color='blue', alpha=0.7)
plt.title("HISTOGRAM OF GOAL DISTRIBUTED TO EACH JERSY NUMBER")
plt.ylabel("Goals per Match")
plt.xlabel("Jersy number")
plt.show()
```

Creating a scater plot for Apperances VS lossers

```python
df = pd.read_csv('C:/Users/user/Desktop/Test 2 prep/dataset - 2020-09-24.csv')
losses = [n for n in df ["Losses"].head(10)]
appearances= [n for n in df ["Appearances"].head(10)]

plt.scatter(appearances,losses)
plt.title("Scatter Plot of Matches lost vs Appearances")
plt.xlabel("Appearances")
plt.ylabel("Losses")
plt.show()
```

Creating a bar chart comparing Names and Appearances for the first six players

```python
players = [n for n in df["Name"].head(6)]
appearances = [n for n in df["Appearances"].head(6)]

plt.figure(figsize=(15, 5))
plt.bar(players, appearances, color='black', alpha=0.7)
plt.title('Game Appearances of each first six Players')
plt.xlabel('Players names')
plt.ylabel('Appearances')
plt.show()
```

```python
creating pie chart to compare goals scored by the first seven players
```

```python
Names = [n for n in df["Name"].head(7)]
goals= [n for n in df["Goals"].head(7)]

plt.figure(figsize=(10,7))
plt.pie(goals, labels=Names, autopct='%1.1f%%', startangle=15)
plt.title("First seven Players's Goals Scored")
plt.show()
```

Box plot

```python
Ycards = [n for n in df["Yellow cards"]]
plt.figure(figsize=(6, 8))
plt.boxplot(Ycards, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title("Distribution of Yellow Cards")
plt.ylabel("Yellow Cards")
plt.grid(True)
plt.show()
```
